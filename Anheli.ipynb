```python

# Python script generated from GoalTracker Text Analysis Report
# You can use this data in environments like Google Colab.

import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud

report_data = {
  "commentDistribution": {
    "activity": "Temprana: El único comentario fue realizado 18 horas después de la publicación inicial, lo que indica interacción en el primer día.",
    "types": [
      "Aportación técnica/complemento: 100%"
    ]
  },
  "executiveSummary": "El post de Alhelí Guadarrama Ponce detalla cómo las computadoras interpretan imágenes a través de Redes Neuronales Convolucionales (CNNs), contrastando la visión humana con la interpretación numérica de las máquinas. Explica las capas clave de una CNN y sus aplicaciones en áreas como vehículos autónomos y diagnósticos médicos. La publicación generó 28 reacciones y un comentario técnico que complementa la explicación.",
  "keyMetrics": {
    "commonEmotionalExpressions": [
      "sorprendentes"
    ],
    "mostFrequentTerms": [
      "CNN",
      "IA",
      "Visión Computacional",
      "Redes Neuronales",
      "Machine Learning",
      "Deep Learning",
      "Data Science"
    ],
    "participationBySector": [
      "Inteligencia Artificial/Data Science/Transformación Digital: 100%"
    ],
    "totalInteractions": "Reactions: 28, Comments: 1"
  },
  "semanticNetwork": "Nodo central: Alhelí Guadarrama Ponce -> explica -> Visión Computacional (a través de CNNs). Conceptos clave: Redes Neuronales, Machine Learning, Deep Learning, IA, Píxeles, Matrices. Aplicaciones mencionadas: Vehículos autónomos, Diagnósticos médicos. El comentario de David Mulato-Martínez complementa el tema principal, añadiendo detalles sobre la percepción de color y vectores.",
  "termEmotionPlot": [
    {
      "emotion": 0.7,
      "frequency": 5,
      "term": "CNN"
    },
    {
      "emotion": 0.6,
      "frequency": 4,
      "term": "IA"
    },
    {
      "emotion": 0.5,
      "frequency": 3,
      "term": "Visión Computacional"
    },
    {
      "emotion": 0.3,
      "frequency": 3,
      "term": "Redes Neuronales"
    },
    {
      "emotion": 0,
      "frequency": 2,
      "term": "Números"
    },
    {
      "emotion": 0.8,
      "frequency": 1,
      "term": "Sorprendentes"
    },
    {
      "emotion": 0.1,
      "frequency": 1,
      "term": "Vectores"
    }
  ],
  "wordCloud": [
    {
      "term": "Visión Computacional",
      "weight": 8
    },
    {
      "term": "IA",
      "weight": 9
    },
    {
      "term": "Computadoras",
      "weight": 6
    },
    {
      "term": "Data Science",
      "weight": 7
    },
    {
      "term": "Números",
      "weight": 5
    },
    {
      "term": "Deep Learning",
      "weight": 7
    },
    {
      "term": "CNN",
      "weight": 10
    },
    {
      "term": "Machine Learning",
      "weight": 7
    },
    {
      "term": "Matrices",
      "weight": 4
    },
    {
      "term": "Redes Neuronales",
      "weight": 8
    },
    {
      "term": "Imágenes",
      "weight": 6
    },
    {
      "term": "Píxeles",
      "weight": 5
    }
  ]
}

def display_report(data):
    """Prints the report data in a readable format."""
    print("--- Executive Summary ---")
    print(data['executiveSummary'])
    print("\n--- Key Metrics ---")
    for key, value in data['keyMetrics'].items():
        print(f"{key}: {value}")
    print("\n--- GoalTracker Semantic Network ---")
    print(data['semanticNetwork'])

def generate_word_cloud(words_data):
    """Generates and displays a word cloud from the report data."""
    if not words_data:
        print("No data for word cloud.")
        return
    word_freq = {item['term']: item['weight'] for item in words_data}
    wc = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.title('GoalTracker  Word Cloud')
    plt.axis('off')
    plt.show()

def generate_term_emotion_chart(chart_data):
    """Generates and displays a bubble chart for term-emotion analysis."""
    if not chart_data:
        print("No data for term-emotion chart.")
        return

    terms = [d['term'] for d in chart_data]
    emotions = [d['emotion'] for d in chart_data]
    frequencies = [d['frequency'] for d in chart_data]

    # Use frequency for bubble size, with some scaling
    sizes = [f * 50 for f in frequencies]

    # Use a jitter for y-axis to avoid overlap
    y_jitter = np.random.rand(len(terms)) * 0.6 - 0.3

    fig, ax = plt.subplots(figsize=(12, 6))

    scatter = ax.scatter(emotions, y_jitter, s=sizes, alpha=0.6, cmap='viridis')

    # Add labels for each bubble
    for i, term in enumerate(terms):
        ax.text(emotions[i], y_jitter[i] + 0.05, term, ha='center', va='bottom', fontsize=9)

    ax.set_title('GoalTracker Term-Emotion Analysis')
    ax.set_xlabel('Sentiment (Negative to Positive)')
    ax.set_xlim(-1.1, 1.1)
    ax.set_xticks([-1, 0, 1])
    ax.set_xticklabels(['Negative', 'Neutral', 'Positive'])

    # Hide y-axis as it's just for spread
    ax.get_yaxis().set_visible(False)

    # Add a vertical line for neutral
    ax.axvline(0, color='grey', linestyle='--', linewidth=0.8)

    plt.grid(True, which='major', axis='x', linestyle='--', linewidth=0.5)
    plt.show()

# --- Main Execution ---
if __name__ == '__main__':
    display_report(report_data)
    print("\n--- GoalTracker Generating Visualizations ---")
    generate_word_cloud(report_data.get('wordCloud'))
    generate_term_emotion_chart(report_data.get('termEmotionPlot'))
```
